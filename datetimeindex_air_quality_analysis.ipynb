```python
import pandas as pd 
import numpy as np 
from datetime import datetime
import os
import matplotlib.pyplot as plt
```

```python
def get_data(dataset,zip_file):
    os.system("unzip datasets/{zip_file}  -d datasets")
    df = pd.read_csv(f"datasets/{dataset}")
    os.system("rm -r datasets/*.csv")
    return df

df = get_data("nyc.csv","airquality_data.zip")
```

```python
# Display information about the dataset before transformations
print(df.info())
```

```python
# Convert the 'date' column from string type to 'datetime64' type
df.date = pd.to_datetime(df.date)
```

```python
# Set the 'date' column as the index of the DataFrame to facilitate time series analysis
df.set_index('date', inplace=True)
```

```python
# Display information about the dataset after setting the date as the index
print(df.info())
```

```python
df.plot(subplots=True)

# Display the plots
plt.show()
```
